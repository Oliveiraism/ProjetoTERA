# Download e extração

```python
#download dos subitens

anos = range(2020,2023)
meses = range(1,13)
url= 'https://ftp.ibge.gov.br/Precos_Indices_de_Precos_ao_Consumidor/IPCA/Resultados_por_Subitem/'
#for ano in anos and for mes in meses:
atual = pd.Timestamp.now()
mesatual = atual.month
anoatual = atual.year
for ano in anos:
  if anoatual != ano:
    for mes in meses:
      print(f' baixando {ano}{"%02d" % mes}')
      download = requests.get(url+f'/{ano}/ipca_{ano}{"%02d" % mes}Subitem.zip')
      print(download)
      # formato do link: https://ftp.ibge.gov.br/Precos_Indices_de_Precos_ao_Consumidor/IPCA/Resultados_por_Subitem/2020/ipca_202001Subitem.zip
      open(f'ipca_{ano}{"%02d" % mes}Subitem.zip', 'wb').write(download.content)
  else:
    for mes in meses:

        if ano == anoatual and mes >= mesatual-1:
          pass
        else:
          print(f' baixando {ano}{"%02d" % mes}')
          download = requests.get(url+f'/ipca_{ano}{"%02d" % mes}Subitem.zip')
          print(download)
          open(f'ipca_{ano}{"%02d" % mes}Subitem.zip', 'wb').write(download.content)

pd.Timestamp.now()
```

```python
veridir()
diretorio_atual = os.getcwd()

for arquivo in os.listdir(diretorio_atual):
  ext = os.path.splitext(arquivo)[-1].lower()

  if ext=='.zip':
    with zipfile.ZipFile(f'{diretorio_atual}/{arquivo}', 'r') as zip_ref:
      zip_ref.extractall(f'{diretorio_atual}')
    print(f'extraindo arquivo{arquivo}')
    
    os.remove(f'{diretorio_atual}/{arquivo}')
  else:
    pass    
```

# Construção da base

```python
#for arquivo in arquivos:
os.getcwd()
agrup= []
seriehistorica = pd.DataFrame(columns= ['legenda', 'RJ', 'POA' ,
  'BH' ,'REC' ,'SP' ,'DF' ,'BEL' ,'FOR' ,'SAL' ,
  'CUR' ,'GOI' ,'VIT' ,'CG' ,'RB' ,'SL' ,'AJU' , 'NACIONAL'])

#arquivo = '/ipca_202106Subitem.xls'
for arquivo in os.listdir(diretorio_atual):
  f = f'{os.getcwd()}/{arquivo}'
  book = xlrd2.open_workbook(f)
  plan = book.sheet_by_index(0)

  legenda = plan.col_values(0)
  RJ = plan.col_values(1)
  POA = plan.col_values(2)
  BH = plan.col_values(3)
  REC = plan.col_values(4)
  SP = plan.col_values(5)
  DF = plan.col_values(6)
  BEL = plan.col_values(7)
  FOR = plan.col_values(8)
  SAL = plan.col_values(9)
  CUR = plan.col_values(10)
  GOI = plan.col_values(11)
  VIT = plan.col_values(12)
  CG = plan.col_values(13)
  RB = plan.col_values(14)
  SL = plan.col_values(15)
  AJU = plan.col_values(16)
  NACIONAL = plan.col_values(17)

  dfa = [legenda[4:-1], RJ[4:-1], POA[4:-1], BH[4:-1], REC[4:-1], 
        SP[4:-1],DF[4:-1],BEL[4:-1],FOR[4:-1], SAL[4:-1], CUR[4:-1],
        GOI[4:-1], VIT[4:-1], CG[4:-1],RB[4:-1],SL[4:-1], AJU[4:-1],
        NACIONAL[4:-1]]
  
  #transposição
  df = pd.DataFrame(dfa).T
  
  df = df.drop(1,axis=0)

  header = df.iloc[0]
  
  df = df[1:]
  df.columns = header

  df['periodo'] = int(arquivo[5:11])

  a=f't{arquivo[5:11]}'
  
  vars()[f't{arquivo[5:11]}'] = df
  agrup.append(df) 

seriehistorica=pd.concat(agrup)
seriehistorica2=pd.DataFrame(seriehistorica)
```
